```python
import numpy as np

class Network(object):
  def __init__(self, size):
    self.size = size
    self.weights = [np.zeros((size[i],size[i+1])) for i in range(len(size)-1)]
    self.bias = [np.zeros((1, size[i+1])) for i in range(len(size)-1)]
    self.initialize_parameters()

  def initialize_parameters(self):
    for i in range(len(self.weights)):
      weight = self.weights[i]
      self.weights[i] = np.random.randn(weight.shape[0], weight.shape[1])
    for i in range(len(self.bias)):
      bias = self.bias[i]
      self.bias[i] = np.random.randn(bias.shape[0])
    
  def feedforward(self,x):
    for i,(weight, bias) in enumerate(zip(self.weights, self.bias)):
      x = weight.T.dot(x) + bias
      x = self.sigmoid(x)
    return x

  def predict(self,x):
    x = self.feedforward(x)
    if x>=0.5:
      return 1
    else:
      return 0
    
  def sigmoid(self, x):
    return 1.0 / (1.0 + np.exp(-x))

  def softmax(self, x):
    x = x - np.max(x) ## nan 해결
    _sum = np.exp(x).sum()
    return np.exp(x)/_sum

def accuracy(preds, targets):
  matrix = []
  for pred, target in zip(preds, targets):
    if pred==target:
      matrix.append(1)
    else:
      matrix.append(0)
  matrix = np.array(matrix)
  return matrix.sum()/len(matrix)
```

```python
size = [4,1]
slp = Network(size)
outputs = []
targets = []
for idx,(train_data, train_label) in enumerate(zip(iris['data'],iris['target'])):
  
  if train_label==2:
    continue
  out = slp.predict(train_data)
  outputs.append(out)
  targets.append(train_label)
  if idx%10==0:
    print("pred : {}, target : {}".format(out, train_label))
print('accuracy : {}'.format(accuracy(outputs, targets)))
print(iris['target'])
```

```python
size = [784,128,32,10]
mlp = Network(size, mode='MNIST')
outputs = []
targets = []
for idx, (train_data, train_label) in enumerate(zip(mnist['data'],mnist['target'])):
  
  pred, pred_index = mlp.predict(train_data)
  outputs.append(pred_index)
  train_label = int(train_label)
  targets.append(train_label)
  if idx%1000==0:
    print("pred : {}, pred_index : {}, target : {}".format(pred, pred_index, train_label))
print('accuracy : {}'.format(accuracy(outputs, targets)))
```
